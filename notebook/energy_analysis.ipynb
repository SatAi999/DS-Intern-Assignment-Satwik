```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append(r'../src')  # Add the src folder to the Python path

# Import the data loading function
from data_loader import load_data

# Load the data using the data loader function
df = load_data(r"C:\Users\sanje\DS-Intern-Assignment-Satwik\data\data.csv")
df
```

```python
# Display all columns and configure visuals
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
%matplotlib inline
```

```python
display(df.head())
display(df.tail())
df.info()
df.describe(include='all')  # Shows stats for both numerical and categorical data
```

```python
# Total missing values per column
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 'Percent (%)': missing_percent})
display(missing_df[missing_df['Missing Values'] > 0].sort_values(by='Percent (%)', ascending=False))
```

```python
# Check data types and unique values for categorical understanding
print(df.dtypes)
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"{col}: {df[col].nunique()} unique values")
```

```python
# Plot histograms of numerical features
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
df[numeric_cols].hist(figsize=(18, 12), bins=30, edgecolor='black')
plt.suptitle("Numerical Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Plot correlation heatmap
plt.figure(figsize=(14, 10))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# Check for outliers in key numeric features
for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
target = 'equipment_energy_consumption'

# Scatterplots for continuous relationships
for col in numeric_cols:
    if col != target:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=df[col], y=df[target])
        plt.title(f"{col} vs {target}")
        plt.show()
```

```python
# Assess usefulness of random variables
sns.pairplot(df[['random_variable1', 'random_variable2', target]])
plt.suptitle("Random Variables vs Target", fontsize=16)
plt.show()
```
